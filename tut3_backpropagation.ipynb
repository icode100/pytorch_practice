here we take x = 1, y=2 and w=1 thus our first gradient should be -2

```python
import torch
x = torch.tensor(1.0)
y = torch.tensor(2.0)
w = torch.tensor(1.0,requires_grad=True)

# forward pass for computing the loss
y_hat = w*x
loss = (y_hat-y)**2
print("loss is",loss)

# backward pass for computing gradient
loss.backward()
print("gradient for the loss wrt weight",w.grad)
```
