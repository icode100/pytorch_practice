### multiplication

````python
x = torch.tensor([[2,3],
                  [4,5]])
y = torch.tensor([[2,2],
                  [9,8]])

torch.mul(x,y) #this is doing element wise multiplication and not matrix multiplication
````

### slicing

````python
x = torch.tensor([[2,3,4,7,4],
                  [4,5,2,3,6],
                  [8,9,1,6,2],
                  [8,9,1,6,2]])
x[:,0] # first element of each row packed into a tensor and returned
````

````python
x[0:2,:]
````

````python
x[0:2,:2]
````

if only one value is present we can convert it into scalar using `.item()`

````python
x[1,2].item()
````

we can transfer operations of tensor from CPU to GPU(if cuda drivers are available) as follows:
```python
if torch.cuda.is_available():
    device = torch.device("cuda")
    x = torch.tensor([[2,3,4,7,4],
                  [4,5,2,3,6],
                  [8,9,1,6,2],
                  [8,9,1,6,2]],device=device)
    # we can also use x.to(device)
    # after this we cannnot simply use x.numpy() as numpy cannot use GPU thus we follow this
    x = x.to("cpu")
    x.numpy()
```
